```python
import re
import pandas as pd
```

```python
f = open('data\_chat.txt','r',encoding='utf-8')
```

```python
data = f.read()
f.close()
print(data)
```

# Spliiting the messages from dates

for this purpose we will be using R.E to split the date and msg

```python
date_pattern = "\[\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{1,2}:\d{1,2}\s(?:PM|AM)\]\s"

messages = re.split(date_pattern,data)[1:]
print(messages)
```

```python
dates = re.findall(date_pattern,data)
print(dates)
```

```python
dates = [d.replace('\u202f', ' ') for d in dates]
print(dates)
```

```python
df = pd.DataFrame({'user_message':messages,'message_date':dates})

df['message_date'] = df['message_date'].str.strip().str.strip('[]')

df['message_date'] = pd.to_datetime(
                                    df['message_date'],
                                    format='%d/%m/%y, %I:%M:%S %p'
                                    )
```

# Splitting User and thier messages

```python
users=[]
messages=[]

for msg in df['user_message']:
    entry = re.split('([\w\W]+?):\s',msg)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages

df.drop(columns=['user_message'],inplace=True)

df.head()
```

```python
df['month'] = df['message_date'].dt.month_name()
df['day'] = df['message_date'].dt.day
df['hour'] = df['message_date'].dt.hour
df['minute'] = df['message_date'].dt.minute

df.head()
```

```python
words = []
for message in df['message']:
    words.extend(message.split())
```
